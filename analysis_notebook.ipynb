```python
x = pd.melt(tourist_df, id_vars='Country', var_name='year_month', value_name='count')
y = x.pivot_table(index='year_month', columns='Country', values='count', aggfunc='sum', fill_value=0)
y.index.name = None
y.columns.name = None
y = y.astype(int)
y['Total'] = y.sum(axis=1)
tourist_df_t = y.reset_index(drop=False).rename(columns={'index':'year_month'})
tourist_df_t.head()
```

We then normalize our tourist dataframe by the population of its respective country. This way each entry will be the proportion of the population that travels into the US for that time frame. We also filtered out irrelevant countries for our analysis, leaving only the countries of interest + total visitors.

```python
# normalize 
population_df = pd.read_csv("datasets/populations.csv").set_index('Country Name')
def process(s):
    year = s.year_month[:4] # gets the year
    if year == '2023':
        year = '2022'

    year_pop = population_df[year]
    def subprocess(i):
        if i == 'year_month':
            return s.year_month
        if i in year_pop.index:
            return s[i]/year_pop[i]
        else:
            return np.nan

    return pd.Series(list(map(subprocess, s.index)), index=s.index)
# apply to all columns
transformed_tourist_df_t = tourist_df_t.apply(process, axis=1)
# only take out the countries of interest
tourist_df_proportions = transformed_tourist_df_t[['Mexico', 'Canada', 'United Kingdom', 'France', 'Total']]
tourist_df_proportions['Total'] = tourist_df_t[['Total','year_month']].apply(lambda x:  x['Total'] / population_df[x['year_month'][:4]].sum() if x['year_month'][:4] != '2023'  else  x['Total'] / population_df['2022'].sum() , axis = 1)
tourist_df_proportions['year_month'] = tourist_df_t['year_month']
```

```python
# get dataframe for only the groups that contain missing values
missing_groups = us_dollar_df[us_dollar_df['dollar_idx'].isnull()]['year_month'].unique()
# take subset
missing_groups = missing_groups[:25]
missing_dollar = us_dollar_df[us_dollar_df['year_month'].apply(lambda x: x in missing_groups)]
# get rid of the zero value to see distribution without nulls right now
missing_dollar = missing_dollar[~missing_dollar['dollar_idx'].isnull()]
# graphing #
before_impute_dist = sns.boxplot(data=missing_dollar, x="dollar_idx", y="year_month", color= 'lightgreen')
before_impute_dist.axes.set_title('Distribution of dollar_idx before imputing missing values (missing values excluded)')
x = plt.yticks(fontsize = 8)
```

We chose to impute these missing values using the mean of the two values before and after the current datapoint (neighbor's means)

```python
# imputes missing values with the mean
missing_idx = list(us_dollar_df[us_dollar_df['dollar_idx'].isnull()].index)

for idx in missing_idx:
    replace = us_dollar_df.loc[idx].copy()
    # day before
    before = us_dollar_df.loc[idx]['time'] - timedelta(days=1)
    # day after
    after = us_dollar_df.loc[idx]['time'] + timedelta(days=1)
    # get next valid previous day before
    day_before = us_dollar_df[us_dollar_df['time'] == before]
    # get next valid day after
    day_after = us_dollar_df[us_dollar_df['time'] == after]
    before_found = False
    after_found = False
    while before_found == False:
        if day_before.shape[0] >= 1:
            dollar_before = day_before['dollar_idx'].iloc[0]
            if math.isnan(dollar_before):
                continue 
            else:
                before_found = True
                break
        before = before - timedelta(days=1)
        day_before = us_dollar_df[us_dollar_df['time'] == before]
   
    while after_found == False:
        if day_after.shape[0] >= 1:
            dollar_after = day_after['dollar_idx'].iloc[0]
            if math.isnan(dollar_after) == False:
                after_found = True 
                break
        after = after + timedelta(days=1)
        day_after = us_dollar_df[us_dollar_df['time'] == after]
    mean = np.mean([dollar_before, dollar_after])
    replace['dollar_idx'] =  mean
    us_dollar_df.loc[idx] = replace

# get data
missing_dollar = us_dollar_df[us_dollar_df['year_month'].apply(lambda x: x in missing_groups)]
# get rid of the zero value to see distribution without nulls right now
missing_dollar = missing_dollar[~missing_dollar['dollar_idx'].isnull()]

# graphing
after_impute_dist = sns.boxplot(data=missing_dollar, x="dollar_idx", y="year_month", color= 'lightgreen')
after_impute_dist.axes.set_title('Distribution of dollar_idx before after imputing missing values')
x = plt.yticks(fontsize = 4)
```

**Checking for `dollar_idx` Autocorrelation**

We tested to see if `dollar_idx` is autocorrelated. Performing an autocorrelation test (Ljung-Box test) we observed that `dollar_idx` is indeed autocorrelated. We decided to tackle this by turning our data into the difference between the current month and the last month, also known as "differencing." This will help to remove the autocorrelation in this feature, which will be useful when modeling something like linear regression.

```python
median_dollars = us_dollar_df.groupby('year_month')[['dollar_idx']].median().reset_index(drop=False)
pval = acorr_ljungbox(median_dollars['dollar_idx'], lags=[12])['lb_pvalue'].iloc[0]
print('pval = ' + str(pval))
if pval < 0.05:
    print('Reject null: residuals are not independent')
else:
    print('Fail to reject null: residuals are independent')
```

```python
# differencing
temp = us_dollar_df.copy()
temp['dollar_idx'] = us_dollar_df['dollar_idx'].diff().diff().dropna()
temp = temp.dropna()
temp = temp.groupby('year_month')[['dollar_idx']].median().reset_index(drop=False).rename(columns={'dollar_idx':'dollar_idx_diff'})
median_dollars = pd.merge(median_dollars, temp, on = 'year_month', how = 'inner')

pval = acorr_ljungbox(median_dollars['dollar_idx_diff'], lags=[12])['lb_pvalue'].iloc[0]
print('pval = ' + str(pval))
if pval < 0.05:
    print('Reject null: residuals are not independent')
else:
    print('Fail to reject null: residuals are independent')
```

Re-doing the test, we can confirm that `dollar_idx_diff` is the not autocorrelated feature version of `dollar_idx`

It is a bit unclear if any of our features are correlated with each othyer, so we used a statistical test to better find out. We look to see if the binary emergency variables are correlated to the continious dollar index variable by calculating the Point-Biserial Correlation Coefficient.

If the Coefficient is:
* -1 = perfect negative correlation between 2 variables
* 0 = no correlation
* +1 = perfect positive correlation between 2 variable

The test also have a p-value calculation. If the p-value is less than our threshold (0.05), then the correlation can be considrfed statistically significant.

```python
conditions = delcaration_type_merged_df.columns[-3:]
count = 0
count_diff = 0

for bev in conditions:
    correlation_coefficient, p_value_idx = pointbiserialr(delcaration_type_merged_df[bev], delcaration_type_merged_df['dollar_idx'])
    correlation_coefficient, p_value_diff = pointbiserialr(delcaration_type_merged_df[bev], delcaration_type_merged_df['dollar_idx_diff'])
    # If p-value lower than threshold, then statistically significant
    if p_value_idx < 0.05:
        count += 1
        print(bev + ' may be correlated with dollar_idx -- Correlation =' + str(correlation_coefficient))

    if p_value_diff < 0.05:
        count_diff += 1
        print(bev + ' may be correlated with dollar_idx -- Correlation =' + str(correlation_coefficient))

print(str(count) + ' condition(s) have been found to be correlated with dollar_idx')
print(str(count_diff) + ' condition(s) have been found to be correlated with dollar_idx_diff')
```

### Quantile Regression


Quantile regression is a statistical modeling technique that builds off of ideas of linear regression to predict conditional quantiles. Quantile regression estimates the conditional median while linear regression estimates the mean. Medians are often less influenced by outliers, making quantile regression a more ideal method to handle datasets with outliers. We used quantile regression because traditional linear regression is often sensitive to outliers, especially in the y-variable. Our y-variable has a few outliers that quantile regression should be able to handle better. Furthermore, quantile regression does not require an assumption of normality in our data distribution. In our case, our data is not perfectly normal so quantile regression may be a better fit than linear regression.

Like linear regression, quantile regression's coefficients are relatively simple to interpret, making it an ideal model for this task.

```python
QUANTILE = 0.5 
# set up variables for analysis
analysis_df = delcaration_type_merged_df.copy()
X = analysis_df[['dollar_idx', 'DR', 'EM','FM']]
y_total = analysis_df['Total']
y_neighbors = analysis_df['neighbors']
y_allies = analysis_df['allies']
```

After inference we looked to see how reliable our models were...

Our test MSE's seem to range from [0.04 ~ 0.06] while our train MSE's range from [0.02 ~ 0.03]. It seems that our models tend to overfit on our training data, which results in almost a 2x worse MSE on our testing data. This indicates that our model capturing noise rather than the pattern we wish to capture in our data.

We unscaled our MSE from the min-max range to the original datarange and turned our MSE into an RMSE value to see how much off our predictions tended to be:

```python
def undo_min_max_scaling(scaled_value, min_val, max_val):
    original_value = (scaled_value * (max_val - min_val)) + min_val
    return original_value
# neighbors
n = undo_min_max_scaling(np.sqrt(test_mse_dict['neighbors']), y_neighbors.min() , y_neighbors.max())
# allies
a = undo_min_max_scaling(np.sqrt(test_mse_dict['allies']), y_allies.min() , y_allies.max())
# total
t = undo_min_max_scaling(np.sqrt(test_mse_dict['total']), y_total.min() , y_total.max())

print('Our neighbors model un-scaled RMSE = {n}\nThe mean of y_neighbors = {nm}\n\n\
Our allies model un-scaled RMSE = {a}\nThe mean of y_neighbors = {am}\n\n\
Our total model un-scaled RMSE = {t}\nThe mean of y_neighbors = {tm}'\
      .format(n=n,a=a,t=t,\
              nm=y_neighbors.mean(), am=y_allies.mean(), tm=y_total.mean()))
```

Looking at our unscaled-RMSE, it seems that our model's predictions tend to be 0.006 off for neighbors, .0014 off for allies, and 2.723e-05 for total.

Comparing this to the original scale (used mean for comparison), it seems that our models tend to be off by a significant amount since our error is a sizable portion of the y-variable's mean in all cases (almost half!). This means that while linear regression does perform better in comparison to other models like quantile regression, it still is not able to perfectly capture a relationship, so we cannot say with strong confidence that our model can accurately capture the relationship between I-94 proportions and our variables. This maybe error from the model overfitting to noise in our training data.

It may also mean either 1) our data was not as good of a fit for linear regression as we thought, hence the high MSE
or 2) our patterns were not as linear as we expected them to be, so linear regression may not be able to truly capture the relationships we wanted.
